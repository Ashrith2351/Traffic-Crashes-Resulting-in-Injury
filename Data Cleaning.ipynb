```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('Traffic_Crashes_Resulting_in_Injury.csv')
data
```

```python
data.columns = data.columns.str.title()
data
```

```python
def title_case_str(val):
    return val.title() if isinstance(val, str) else val
data = data.applymap(title_case_str)
data
```

```python
#Checking for missing values
completeness_check = data.isnull().sum()
print("Completeness Check:")
print(completeness_check)
```

```python
null_check = data[data.isnull().any(axis=1)]
print("\nRows with NULL values:")
print(null_check)
```

```python
duplicate_check = data[data.duplicated(keep='first')]
print("\nDuplicate Records:")
print(duplicate_check)
```

```python
data = data.dropna()
data
```

```python
data = data.reset_index(drop=True)
data
```

```python
summary_stats = data.describe()
print(summary_stats)
```

```python
data = data.drop('Weather_2', axis = 1)
data
```

```python
data.to_csv('Traffic_Crashes_Resulting_in_Injury_Cleaned.csv', index=False)
```
